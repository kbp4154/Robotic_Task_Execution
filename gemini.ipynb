```python
def read_file_as_string(filename):
  with open(filename, "r") as f:
    content = f.read()
  return content

def read_json(filename):
  with open(filename, 'r') as file:
      data = json.load(file)
  return data

def save_task_tree_as_json(text,dish):
   try:
    task_trees = json.loads(text)
    if not os.path.exists(f'/content/drive/My Drive/DL Project 3/prompt_outputfiles/{dish}'):
      os.makedirs(f'/content/drive/My Drive/DL Project 3/prompt_outputfiles/{dish}')
    for i, task_tree in enumerate(task_trees, start=1):
      filename = os.path.join(f'/content/drive/My Drive/DL Project 3/prompt_outputfiles/{dish}/{dish}_{i}.json')
      with open(filename, 'w') as file:
        json.dump(task_tree, file, indent=2)
   except Exception as e:
    with open('/content/drive/My Drive/DL Project 3/prompt_outputfiles/error.txt', 'w') as et:
      et.write(text)
    print(e)
    exit()

def save_task_tree_as_json2(text,dish):
   try:
    task_trees = json.loads(text)
    if not os.path.exists(f'/content/drive/My Drive/DL Project 3/prompt2_outputfiles/{dish}'):
      os.makedirs(f'/content/drive/My Drive/DL Project 3/prompt2_outputfiles/{dish}')
    for i, task_tree in enumerate(task_trees, start=1):
      filename = os.path.join(f'/content/drive/My Drive/DL Project 3/prompt2_outputfiles/{dish}/{dish}_{i}.json')
      with open(filename, 'w') as file:
        json.dump(task_tree, file, indent=2)
   except Exception as e:
    with open('/content/drive/My Drive/DL Project 3/prompt2_outputfiles/error.txt', 'w') as et:
      et.write(text)
    print(e)
    exit()


def save_task_tree_as_json3(text,dish):
   try:
    task_trees = json.loads(text)
    if not os.path.exists(f'/content/drive/My Drive/DL Project 3/prompt3_outputfiles/{dish}'):
      os.makedirs(f'/content/drive/My Drive/DL Project 3/prompt3_outputfiles/{dish}')
    for i, task_tree in enumerate(task_trees, start=1):
      filename = os.path.join(f'/content/drive/My Drive/DL Project 3/prompt3_outputfiles/{dish}/{dish}_{i}.json')
      with open(filename, 'w') as file:
        json.dump(task_tree, file, indent=2)
   except Exception as e:
    with open('/content/drive/My Drive/DL Project 3/prompt3_outputfiles/error.txt', 'w') as et:
      et.write(text)
    print(e)
    exit()
```

```python
if __name__ == "__main__":
   read_json_file = read_json('/content/drive/My Drive/DL Project 3/input.json')
   dishes_skipped = []
   for data in  read_json_file:
      dish_category,dish_menu = data['category'],data['menu']
      for dish_menu_data in dish_menu:
         dish, ingredients = dish_menu_data['dish'], dish_menu_data['ingredients']
         print("Dish: ", dish, ", Ingredients: ", ingredients,'\n')
        #  read_prompt = read_file_as_string("/content/drive/My Drive/DL Project 3/prompt1.txt")
        #  read_prompt = read_file_as_string("/content/drive/My Drive/DL Project 3/prompt2.txt")
         read_prompt = read_file_as_string("/content/drive/My Drive/DL Project 3/prompt3.txt")
         read_prompt = read_prompt.replace('{dish}', dish)
         read_prompt = read_prompt.replace('{ingredients}', str(ingredients))
         print(read_prompt.split('\n')[0])
        #  error_count = 0
        #  text_model = genai.GenerativeModel('gemini-1.5-pro-latest')
         text_model = genai.GenerativeModel('gemini-pro')
         try:
          response = text_model.generate_content(read_prompt)
          json.loads(response.text)
          # save_task_tree_as_json(response.text, dish)
          # save_task_tree_as_json2(response.text, dish)
          save_task_tree_as_json3(response.text, dish)
         except:
          dishes_skipped.append([dish, ingredients])
          print("Encountered an error:-----------------------------------", dish, ingredients)

   print("\n")
   print("dishes skipped were :- ", dishes_skipped)
```
